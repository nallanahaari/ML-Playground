```python
from sklearn.model_selection import train_test_split as tts
xtrain, xtest, ytrain, ytest = tts(x,y,test_size=0.20, random_state=42)
```

```python
xtrain
```

```python
from sklearn.linear_model import LogisticRegression as lor
reg = lor()
```

```python
from sklearn.model_selection import  GridSearchCV
```

```python
parameter = {'C':[0.5,1,2,3,5,7,9,10,40,50], 'penalty':['l1','l2','elasticnet']}
log_regressor = GridSearchCV(reg,parameter,scoring='accuracy',cv=5)
log_regressor.fit(x,y)
```

```python
print(log_regressor.best_params_)
```

```python
print(log_regressor.best_score_)
```

```python
prediction_log = log_regressor.predict(xtest)
```
